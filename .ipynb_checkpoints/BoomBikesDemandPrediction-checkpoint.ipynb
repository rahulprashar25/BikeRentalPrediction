# 3. Data Preparation

### Creating Dummy variables

```python
#Year can be changed to 0 and 1 
day1 = day
day1['yr'] = day1.yr.map({'2018':0,'2019':1})
day1.head()
```

```python
#Dummy variables for season,mnth, weekday,weathersit

season = pd.get_dummies(day1.season,drop_first=True)
weekday = pd.get_dummies(day1.weekday,drop_first=True)
weathersit = pd.get_dummies(day1.weathersit,drop_first=True)
mnth = pd.get_dummies(day1.mnth,drop_first=True)

```

```python
season.head()
```

```python
weekday.head()
```

```python
weathersit.head()
```

```python
mnth.head()
```

* Dummy variables are less than 1 
* It can be now concatenated with the orginal day1 dataset 
* We can then remmove these fields from original dataset

```python
day1 = pd.concat([day1,season,weekday,weathersit,mnth],axis = 1)
day1.head()
```

```python
#Drop  the categorical variables now

day2 = day1.drop(['mnth','weekday','weathersit','season'],axis=1)

day2.head()
```


### Splitting the data into test train 

```python
df_train, df_test = train_test_split(day2,train_size = 0.7, random_state = 100)
print(df_train.shape)
print(df_test.shape)
```

### Rescaling feautures using Min-Max Scaling

```python
#Scaling numeric variable
scaler = MinMaxScaler()
num_vars = ['temp','hum','windspeed','cnt']


df_train[num_vars] = scaler.fit_transform(df_train[num_vars])
df_train.head()
```

# 4. Building the model

### Training the model

```python
#Heatmap of the new data

plt.figure(figsize = (36,15))
sns.heatmap(df_train.corr(),annot = True, cmap = "YlGnBu")
plt.show()
```

```python
#X_train, y_train
y_train = df_train.pop('cnt')
X_train = df_train
```

### Using statsmodel

```python
#Building model with all the variables

X_train_sm = sm.add_constant(X_train)

#Create Model
lr = sm.OLS(y_train,X_train_sm)

#fit
lr_model = lr.fit()

#params
lr_model.summary()
```

### Checking VIF

```python
vif = pd.DataFrame()
vif['Features'] =  X_train.columns
vif['VIF'] = [variance_inflation_factor(X_train.values,i) for i in range(X_train.shape[1])]
vif['VIF'] = round(vif['VIF'],2)
vif = vif.sort_values(by = "VIF",ascending = False)
vif
```

```python
#Dropping high p-values
X = X_train.drop(['holiday','Wed','June','Mar','Oct','Aug','Thu'],axis = 1)
```

```python
#Building another model with dropped variables

X_train_sm = sm.add_constant(X)

#Create Model
lr = sm.OLS(y_train,X_train_sm)

#fit
lr_model = lr.fit()

#params
lr_model.summary()
```

```python
vif = pd.DataFrame()
vif['Features'] =  X.columns
vif['VIF'] = [variance_inflation_factor(X.values,i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'],2)
vif = vif.sort_values(by = "VIF",ascending = False)
vif
```

```python
#Dropping high p-values
X1 = X.drop(['Sun','Tue','summer','Feb','May','Mon'],axis = 1)
```

```python
#Building another model with dropped variables

X_train_sm = sm.add_constant(X1)

#Create Model
lr = sm.OLS(y_train,X_train_sm)

#fit
lr_model = lr.fit()

#params
lr_model.summary()
```

```python
vif = pd.DataFrame()
vif['Features'] =  X1.columns
vif['VIF'] = [variance_inflation_factor(X1.values,i) for i in range(X1.shape[1])]
vif['VIF'] = round(vif['VIF'],2)
vif = vif.sort_values(by = "VIF",ascending = False)
vif
```

```python
#Drop humidity and rebuild the model
X2 = X1.drop('hum',axis = 1)
```

```python
#Building another model with dropped variables

X_train_sm = sm.add_constant(X2)

#Create Model
lr = sm.OLS(y_train,X_train_sm)

#fit
lr_model = lr.fit()

#params
lr_model.summary()
```

```python
vif = pd.DataFrame()
vif['Features'] =  X2.columns
vif['VIF'] = [variance_inflation_factor(X2.values,i) for i in range(X2.shape[1])]
vif['VIF'] = round(vif['VIF'],2)
vif = vif.sort_values(by = "VIF",ascending = False)
vif
```

### Now all the features looks in the range we can do Residual Analysis and Prediction

# 5. Residual Analysis

```python
y_train_pred = lr_model.predict(X_train_sm)
```

```python
res = y_train - y_train_pred
sns.distplot(res)
```

**Error is normally distributed**

### Prediction and evaluation of the test set

```python

num_vars = ['temp','hum','windspeed','cnt']


df_test[num_vars] = scaler.transform(df_test[num_vars])
df_test.head()
```

```python
df_test.describe()
```

```python
y_test = df_test.pop('cnt')
X_test = df_test
```

```python
# add a constant to X_test
X_test_sm = sm.add_constant(X_test)
X_test_sm.head()
```

```python
#Drop the variables dropped in train test
X_test_sm = X_test_sm.drop(['Sun','Tue','summer','Feb','May','Mon','holiday','Wed','June','Mar','Oct','Aug','Thu','hum'],axis =1)
```

```python
#Predict 
y_test_pred = lr_model.predict(X_test_sm)
```

```python
#Evaluate
#r2_score test data

r2_score(y_true = y_test, y_pred = y_test_pred)
```

```python
#r2_score train data
r2_score(y_true = y_train, y_pred = y_train_pred)
```

### Using RFE to find top 3 features

```python
#Import RFE

from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
lm= LinearRegression()
lm.fit(X_train,y_train)
rfe = RFE(lm,3)
rfe = rfe.fit(X_train,y_train)
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

**Temperature, humidity & Windspeed are top 3 variables**

### Checking Homoscedasticity

```python
check_hetero = pd.DataFrame()
```

```python
check_hetero['res']  = res
check_hetero['y_train_pred']  = y_train_pred
check_hetero
```

```python
sns.scatterplot(x='y_train_pred',y='res',data = check_hetero)
plt.show()
```

**We can see that the residuals have similar variance for different predictied values of Y, thus the 3rd assumption of Homoscedasticity is correct**

### Result
**We can see from the coefficients of the test and train data calculated above, that temperature is the major factor in predicting the demand for Bike Rentals**
